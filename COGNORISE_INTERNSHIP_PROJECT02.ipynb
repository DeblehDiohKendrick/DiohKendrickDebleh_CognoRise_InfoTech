### UNEMPLOYMENT IN INDIA
Question: UNEMPLOYMENT is measured by the Unemployment Rate, which is the Number of People who are skilled, but Unemployed as a Percentage of the Total Labor Force. There is a sharp increase in Unemployment Rate during the Covid19 Pandemic in India. Therefore, this analysis provides insights to socio-economic challenges Covid19 Pademic imposed on Indian workforce and the labor market.

To conduct this analysis, we use the UNEMPLOYMENT IN INDIA dataset, which dynamically covers the entire country (India), during the Covid19 pandemic. This data provides information on the impact of unemployment (negatively) and how the labor market dropped due to unemployment.

```python
Data = pd.read_csv('CognoRise_Unemployment in India.csv')
Data
```

### Using the isnull() method to check fields which are empty. Note, this returns a Boolean value (True - which means cell is empty, and False - which means there is data in the cell)

```python
Data.isnull()
```

### Determining the isnull() cells, sum them up to know the total by using the 'isnull().sum()' method.

```python
Data.isnull().sum()
```

### Drop the null cells by using the 'dropna()' method.

```python
Data.dropna(inplace=True)
```

### Apply the 'shape' method to get the new shape of the data after we have dropped unnecessary cells. This process is for cleaning the data for accuracy and uniqueness.

```python
Data.shape
```

### Load the data. The rows of data reduced to 740, maintaining the columns.

```python
Data
```

### Re-run the isnull() method to prove all fields have records, giving us 'False'.

```python
Data.isnull()
```

### Rename columns to maximize space to view entire data.

```python
Data.columns=['Region','Date','Frequency','Estimated_Emp_Rate(%)','Estimated_Emp','Estimated_Labor_Rate(%)','Area']
print(Data)
```

### After renaming columns, load the data to see the effect.

```python
Data
```

### 'Estimated_Emp' column has large figures. Therefore, we use the 'round' method to convert them into Indian numbering (Lahks). Example, where we had 11999139.0, having applied the 'round' method, we now have 12.00 Lakhs (100,000 per Lakh)

```python
Data['Estimated_Emp'] = round((Data['Estimated_Emp']/1000000),2)
Data
```

### Convert the 'Date' column to 'Datetime'.

```python
Data['Date'] = pd.to_datetime(Data['Date'])
Data.info()
```

### Apply the duplicated().sum() methods to determine duplicate values and their total.

```python
Data.duplicated().sum()
```

### Apply the value_counts() method on the 'Frequency' column. 

```python
Data.Frequency.value_counts()
```

### This analysis does not require the 'Frequency' column. Therefore, use the drop() method to delete. The result gives 740 rows and 6 columns. Good job.

```python
Data = Data.drop(['Frequency'], axis=1)
Data.shape
```

### Initially changed the 'Date' column to Datetime. Now, separate and have individual columns for 'Year' and 'Month'. This adds two separate columns to the data, giving 740 rows and 8 columns.

```python
Data['Year'] = Data['Date'].dt.year
Data['Month'] = Data['Date'].dt.month
Data
```

### Use the 'value_counts() to normalize our data in the 'Area' column.

```python
Data.Area.value_counts(normalize=True)*100
```

### Now, begin plotting and displaying visuals.

### 'Countplot' of Area, which provides the total for Urban and Rural areas.

```python
plt.figure(figsize = [6,5])
ax = sns.countplot(data = Data, x = 'Area')

plt.title("Countplot of Area", fontdict={'fontsize': 16, 'fontweight' : 12, 'color' : 'Black'})
for x in ax.containers:
        ax.bar_label(x,rotation=0)

plt.show()
```

### Use the 'value_counts()' to determine all the Regions in India.

```python
Data.Region.value_counts()
```

### This countplot provides details per Region and Area, with a legend applied in the upper right corner. It tells the total per region. As seen, Chandigarh does not have a Rural Area, rather Urban Area only.

```python
plt.figure(figsize=(12, 6))
plt.style.use('fivethirtyeight')
ax = sns.countplot(x=Data['Region'], hue= Data['Area'], palette=['red','blue'])
plt.legend(loc = 'upper right')
for x in ax.containers:
    ax.bar_label(x, rotation=75)
plt.xticks(rotation=75)
plt.ylim(0, 18)
plt.title('Countplot of Region with Overview of Area')
plt.show()
```

### Use the 'ggplot' to measure unemployment between 2019 and 2020. From this analysis, Unemployment Rate was extremely high in 2020 compared to 2019.

```python
plt.figure(figsize=(8,8))
plt.style.use('ggplot')
sns.boxplot(y=Data['Estimated_Emp_Rate(%)'], x=Data['Year'], color='blue')
plt.ylabel('Unemployment Rate in India')
plt.show()
```

### This visual gives the yearly unemployment rate from 2018 - 2020, with each divided into quarters.

```python
plt.figure(figsize=(6,4))
plt.style.use('ggplot')
sns.scatterplot(x=Data['Estimated_Emp_Rate(%)'], y= Data['Year'])
plt.ylim(2018, 2020)
plt.title('Unemployment Rate per Year')
plt.show()
```

### This visual provides the Labor Rate per Region, with Telangana surpassing all regions.

```python
plt.figure(figsize=(12, 8))
plt.style.use('ggplot')
sns.scatterplot(x=Data['Estimated_Labor_Rate(%)'], y=Data['Region'], color='blue')
plt.title('Labor Rate per Region')
plt.show()
```

### We find the correlation.

```python
numeric_cols = [col for col in Data.columns if Data[col].dtype != object]
corr_matrix = Data[numeric_cols].corr()
plt.figure(figsize=(5, 3))
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

### Finding the Percentage of Labor Participation Rate Yearly

```python
sns.lineplot(data=Data, x='Date', y='Estimated_Labor_Rate(%)', hue='Area', errorbar=None, size=0.5)
plt.title('Percentage of Labor Participation Rate Yearly')
plt.ylabel('Labor Participation Rate')
plt.xticks(rotation=80)
plt.show()
```

To conclude, UNEMPLOYMENT RATE increased from 35% in 2019 to a bit over 76% in 2020.
