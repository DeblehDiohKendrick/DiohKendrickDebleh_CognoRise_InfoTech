### Import Python Packages for Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso
```

### Upload CognoRise_Cereal Dataset

```python
Cereal = pd.read_csv('CognoRise_Cereal.csv')
Cereal
```

## Get Info about the Dataset

```python
Cereal.info()
```

### Describe the Dataset

```python
Cereal.describe
```

### Get Sum of -1 

```python
(Cereal == -1).sum()
```

### Drop 'name' column as it is not useful for this analysis.

```python
Cereal = Cereal.drop('name', axis=1)
```

### Dispay the data after dropping the 'Name' column

```python
Cereal
```

### Replace -1 with NaN using Numpy.

```python
Cereal = Cereal.replace(-1, np.NaN)
```

### Use the .isna().sum() methods to determine the null fields in the data. This showed that carbo, sugars and potass have null fields.

```python
Cereal.isna().sum()
```

### Iterate over columns to fill with fillna (mean of each)

```python
for column in ['carbo','sugars','potass']:
    Cereal[column] = Cereal[column].fillna(Cereal[column].mean())
```

```python
Cereal.isna().sum()
```

### Display the data after applying the fillna() method.

```python
Cereal
```

### Convert the 'mfr','type' columns into list for smooth analysis.

```python
{column: list(Cereal[column].unique()) for column in ['mfr','type']}
```

```python
Cereal['type'] = Cereal['type'].apply(lambda x: 1 if x =='H' else 0)
```

### Create a DataFrame using Dummies for 'mfr' column to represent categorical data as numerical values

```python
dummies = pd.get_dummies(Cereal['mfr'])
Cereal = pd.concat([Cereal, dummies], axis=1)
Cereal = Cereal.drop('mfr', axis=1)
```

### Display the data after applying 'Dummies'.

```python
Cereal
```

### Split & Scale

```python
y = Cereal.loc[:, 'rating']
X = Cereal.drop('rating', axis=1)
X
```

### Displaying 'y' data after the drop() method to delete 'rating' column

```python
y
```

### Create the Scaler from the Scikit-Learn, a Preprocessing technique to standardize features by removing unit variance.

```python
scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
X
```

### Training & Testing the data. Training data takes more, 70% of the data, while the test takes 30%, using a random scale of 42.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)
```

### Training suing LinearRegression, Ridge and Lasso

```python
model = LinearRegression()
l2_model = Ridge(alpha=1.5)
l1_model = Lasso(alpha=0.001)
```

### Apply the fit() method to determine whether the data fit the model.

```python
model.fit(X_train, y_train)
l2_model.fit(X_train, y_train)
l1_model.fit(X_train, y_train)

print("Models trained.")
```

### Restore Results on how fit data is after fitting the model to the data.

```python
model_r2 = model.score(X_test, y_test)
l2_model_r2 = l2_model.score(X_test, y_test)
l1_model_r2 = l1_model.score(X_test, y_test)
```

### The R^2 Score evaluates the performance of Regression Models

```python
print("R^2 Scores\n" + "*" * 10)
print("        Without Regularization:", model_r2)
print("With L2 (Ridge) Regularization:", l2_model_r2)
print("With L1 (Lasso) Regularization:", l1_model_r2)
```
